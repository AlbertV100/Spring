```python
def run_sql_file(db_file, sql_file):
    conn = sqlite3.connect(db_file)
    cur = conn.cursor()

    with open(sql_file, 'r') as f:
        sql_script = f.read()

    cur.executescript(sql_script)
    conn.commit()
    conn.close()
```

/* Q1: Some of the facilities charge a fee to members, but some do not.
Write a SQL query to produce a list of the names of the facilities that do. */


```python
import sqlite3

# Connect once at the start
conn = sqlite3.connect(r"C:\Springboard\SQL and Databases\8.3 Case Study\1602668743_SQLFiles_Tier_1_updated_10_13_2020\SQLFiles Tier 1\sqlite_db_pythonsqlite.db")
cur = conn.cursor()

# Q1: Facilities that charge a fee to members
q1 = """
SELECT name
FROM Facilities
WHERE membercost > 0;
"""

cur.execute(q1)
rows = cur.fetchall()

print("Facilities that charge a fee to members:")
for row in rows:
    print(row[0])
```

/* Q2: How many facilities do not charge a fee to members? */

```python
q2 = """
SELECT COUNT(*) AS free_facilities
FROM Facilities
WHERE membercost = 0;
"""

result = run_query(q2)
print("Number of facilities that do not charge a fee:", result[0][0])
```

/* Q3: Write an SQL query to show a list of facilities that charge a fee to members,
where the fee is less than 20% of the facility's monthly maintenance cost.
Return the facid, facility name, member cost, and monthly maintenance of the
facilities in question. */

```python
q3 = """
SELECT facid, name, membercost, monthlymaintenance
FROM Facilities
WHERE membercost > 0
  AND membercost < 0.2 * monthlymaintenance;
"""

results = run_query(q3)

print("Facilities where member fee < 20% of monthly maintenance:")
for row in results:
    print(row)
```

/* Q4: Write an SQL query to retrieve the details of facilities with ID 1 and 5.
Try writing the query without using the OR operator. */

```python
q4 = """
SELECT *
FROM Facilities
WHERE facid IN (1, 5);
"""

results = run_query(q4)

print("Facilities with ID 1 and 5:")
for row in results:
    print(row)
```

/* Q5: Produce a list of facilities, with each labelled as
'cheap' or 'expensive', depending on if their monthly maintenance cost is
more than $100. Return the name and monthly maintenance of the facilities
in question. */

```python
q5 = """
SELECT name,
       monthlymaintenance,
       CASE 
           WHEN monthlymaintenance > 100 THEN 'expensive'
           ELSE 'cheap'
       END AS cost_label
FROM Facilities;
"""

results = run_query(q5)

# Convert to DataFrame for nicer display
df_q5 = pd.DataFrame(results, columns=["name", "monthlymaintenance", "cost_label"])
display(df_q5)
```

/* Q7: Produce a list of all members who have used a tennis court.
Include in your output the name of the court, and the name of the member
formatted as a single column. Ensure no duplicate data, and order by
the member name. */

```python
q7 = """
SELECT DISTINCT
       f.name AS court,
       m.firstname || ' ' || m.surname AS member
FROM Bookings b
JOIN Facilities f USING (facid)
JOIN Members m USING (memid)
WHERE f.name LIKE 'Tennis Court%' AND b.memid <> 0
ORDER BY member;
"""
rows = run_query(q7)
df_q7 = pd.DataFrame(rows, columns=["court", "member"])
display(df_q7)
```

/* Q8: Produce a list of bookings on the day of 2012-09-14 which
will cost the member (or guest) more than $30. Remember that guests have
different costs to members (the listed costs are per half-hour 'slot'), and
the guest user's ID is always 0. Include in your output the name of the
facility, the name of the member formatted as a single column, and the cost.
Order by descending cost, and do not use any subqueries. */

```python
q8 = """
SELECT
  f.name AS facility,
  CASE WHEN m.memid = 0 THEN 'Guest'
       ELSE m.firstname || ' ' || m.surname END AS member,
  (b.slots * CASE WHEN b.memid = 0 THEN f.guestcost ELSE f.membercost END) AS cost
FROM Bookings b
JOIN Facilities f USING (facid)
JOIN Members m USING (memid)
WHERE date(b.starttime) = '2012-09-14'
  AND (b.slots * CASE WHEN b.memid = 0 THEN f.guestcost ELSE f.membercost END) > 30
ORDER BY cost DESC;
"""
rows = run_query(q8)
df_q8 = pd.DataFrame(rows, columns=["facility", "member", "cost"])
display(df_q8)
```

/* Q9: This time, produce the same result as in Q8, but using a subquery. */

```python
q9 = """
SELECT facility, member, cost
FROM (
  SELECT
    f.name AS facility,
    CASE WHEN m.memid = 0 THEN 'Guest'
         ELSE m.firstname || ' ' || m.surname END AS member,
    (b.slots * CASE WHEN b.memid = 0 THEN f.guestcost ELSE f.membercost END) AS cost
  FROM Bookings b
  JOIN Facilities f USING (facid)
  JOIN Members m USING (memid)
  WHERE date(b.starttime) = '2012-09-14'
) AS x
WHERE cost > 30
ORDER BY cost DESC;
"""
rows = run_query(q9)
df_q9 = pd.DataFrame(rows, columns=["facility", "member", "cost"])
display(df_q9)
```

/* Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members! */

```python
q10 = """
SELECT
  f.name AS facility,
  SUM(b.slots * CASE WHEN b.memid = 0 THEN f.guestcost ELSE f.membercost END) AS revenue
FROM Bookings b
JOIN Facilities f USING (facid)
GROUP BY f.facid, f.name
HAVING SUM(b.slots * CASE WHEN b.memid = 0 THEN f.guestcost ELSE f.membercost END) < 1000
ORDER BY revenue;
"""
rows = run_query(q10)
df_q10 = pd.DataFrame(rows, columns=["facility", "revenue"])
display(df_q10)
```

/* Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order */


```python
q11 = """
SELECT
  m.surname || ', ' || m.firstname AS member,
  COALESCE(r.surname || ', ' || r.firstname, '') AS recommended_by
FROM Members m
LEFT JOIN Members r ON m.recommendedby = r.memid
ORDER BY m.surname, m.firstname;
"""
rows = run_query(q11)
df_q11 = pd.DataFrame(rows, columns=["member", "recommended_by"])
display(df_q11)
```

/* Q12: Find the facilities with their usage by member, but not guests */

```python
q12 = """
SELECT
  f.name AS facility,
  m.surname || ', ' || m.firstname AS member,
  SUM(b.slots) AS slots
FROM Bookings b
JOIN Facilities f USING (facid)
JOIN Members m USING (memid)
WHERE b.memid <> 0
GROUP BY f.name, m.memid
ORDER BY f.name, member;
"""
rows = run_query(q12)
df_q12 = pd.DataFrame(rows, columns=["facility", "member", "slots"])
display(df_q12)
```

/* Q13: Find the facilities usage by month, but not guests */

```python
q13 = """
SELECT
  f.name AS facility,
  strftime('%Y-%m', b.starttime) AS month,
  SUM(b.slots) AS slots
FROM Bookings b
JOIN Facilities f USING (facid)
WHERE b.memid <> 0
GROUP BY f.name, month
ORDER BY f.name, month;
"""
rows = run_query(q13)
df_q13 = pd.DataFrame(rows, columns=["facility", "month", "slots"])
display(df_q13)
```
