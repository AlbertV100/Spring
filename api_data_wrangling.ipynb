```python
# Import Libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Optional: Configure matplotlib for inline display in Jupyter
%matplotlib inline
```

## Understand the Open-Meteo API

- **API Endpoint**: `https://archive-api.open-meteo.com/v1/archive`
- **Parameters**:
  - `latitude`: Latitude of the location.
  - `longitude`: Longitude of the location.
  - `start_date`: Start date of the data (YYYY-MM-DD).
  - `end_date`: End date of the data (YYYY-MM-DD).
  - `hourly`: Comma-separated list of variables (e.g., temperature, precipitation).
  - `timezone`: Timezone of the data.

**Documentation**: [Open-Meteo API Docs](https://open-meteo.com/en/docs)

```python
# Step 1: Set Up Parameters

# Location coordinates for New York City
latitude = 40.7128
longitude = -74.0060

# Date range
start_date = '2023-01-01'
end_date = '2023-01-07'  # One week of data

# Variables to fetch
hourly_variables = ['temperature_2m', 'relativehumidity_2m', 'precipitation']
```

```python
# Step 2: Make the API Request

base_url = 'https://archive-api.open-meteo.com/v1/archive'

params = {
    'latitude': latitude,
    'longitude': longitude,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/New_York'
}

response = requests.get(base_url, params=params)
```

```python
# Check if the request was successful

if response.status_code == 200:
    print('Data fetched successfully!')
else:
    print(f'Failed to fetch data. Status code: {response.status_code}')
```

```python
# Step 3: Load Data into pandas DataFrame

data = response.json()
hourly_data = data['hourly']

df = pd.DataFrame(hourly_data)
df.head()
```

```python
# Step 4: Data Cleaning

# Convert 'time' column to datetime
df['time'] = pd.to_datetime(df['time'])

# Set 'time' as the index
df.set_index('time', inplace=True)

# Check for missing values
df.isnull().sum()
```

```python
# Handle missing values (if any)
df.fillna(method='ffill', inplace=True)  # Forward fill
```

```python
# Step 5: Exploratory Data Analysis

# Summary Statistics
df.describe()
```

```python
# Plot Temperature Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['temperature_2m'], label='Temperature (°C)')
plt.title('Temperature Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
# Plot Relative Humidity Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['relativehumidity_2m'], color='orange', label='Relative Humidity (%)')
plt.title('Relative Humidity Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

```python
# Plot Precipitation Over Time

plt.figure(figsize=(14, 6))
plt.bar(df.index, df['precipitation'], label='Precipitation (mm)')
plt.title('Precipitation Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

```python
# Step 6: Correlation Analysis

# Calculate correlation between temperature and humidity
correlation = df['temperature_2m'].corr(df['relativehumidity_2m'])
print(f'Correlation between Temperature and Relative Humidity: {correlation:.2f}')
```

```python
# Scatter Plot of Temperature vs. Relative Humidity

plt.figure(figsize=(8, 6))
plt.scatter(df['temperature_2m'], df['relativehumidity_2m'], alpha=0.5)
plt.title('Temperature vs. Relative Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
# Step 7: Resample Data to Daily Averages

daily_avg = df.resample('D').mean()
daily_avg.head()
```

```python
# Plot Daily Average Temperature

plt.figure(figsize=(10, 5))
plt.plot(daily_avg.index, daily_avg['temperature_2m'], marker='o')
plt.title('Daily Average Temperature in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
# Step 8: Compare with Another Location (Los Angeles)

# Coordinates for Los Angeles
latitude_la = 34.0522
longitude_la = -118.2437

params_la = {
    'latitude': latitude_la,
    'longitude': longitude_la,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/Los_Angeles'
}

response_la = requests.get(base_url, params=params_la)

# Load and clean Los Angeles data
data_la = response_la.json()
hourly_data_la = data_la['hourly']
df_la = pd.DataFrame(hourly_data_la)

df_la['time'] = pd.to_datetime(df_la['time'])
df_la.set_index('time', inplace=True)
df_la.fillna(method='ffill', inplace=True)
```

```python
# Resample Los Angeles Data to Daily Averages

daily_avg_la = df_la.resample('D').mean()
```

```python
# Combine DataFrames for Comparison

combined_temp = pd.DataFrame({
    'New York': daily_avg['temperature_2m'],
    'Los Angeles': daily_avg_la['temperature_2m']
})
```

```python
# Plot Comparison of Daily Average Temperature

combined_temp.plot(kind='bar', figsize=(10, 6))
plt.title('Daily Average Temperature: New York vs Los Angeles')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Step 9: Advanced Analysis - Temperature Difference

# Calculate Temperature Difference
combined_temp['Temp Difference'] = combined_temp['Los Angeles'] - combined_temp['New York']
combined_temp
```

```python
# Step 10: Save Data to CSV (Optional)

# Save New York data
df.to_csv('new_york_weather.csv')

# Save Los Angeles data
df_la.to_csv('los_angeles_weather.csv')
```

```python
# Step 1: Set Up Parameters (London & Seattle, 2022)

# London
lon_latitude = 51.5074
lon_longitude = -0.1278
lon_timezone = 'Europe/London'

# Seattle
sea_latitude = 47.6062
sea_longitude = -122.3321
sea_timezone = 'America/Los_Angeles'

# Full year 2022
start_date = '2022-01-01'
end_date   = '2022-12-31'

# Variables to fetch (hourly precipitation)
hourly_variables = ['precipitation']
```

```python
# Define the simple fetch function
import requests

def fetch_city_hourly_prec(lat, lon, start_date, end_date, tz):
    base_url = 'https://archive-api.open-meteo.com/v1/archive'
    params = {
        'latitude': lat,
        'longitude': lon,
        'start_date': start_date,
        'end_date': end_date,
        'hourly': 'precipitation',
        'timezone': tz
    }
    resp = requests.get(base_url, params=params)
    if resp.status_code != 200:
        raise RuntimeError(f"Request failed ({resp.status_code}): {resp.text[:300]}")
    data = resp.json()
    if 'hourly' not in data or 'precipitation' not in data['hourly']:
        raise ValueError("Hourly precipitation not found in response.")
    return data['hourly']
```

```python
lon_hourly = fetch_city_hourly_prec(lon_latitude, lon_longitude, start_date, end_date, lon_timezone)
sea_hourly = fetch_city_hourly_prec(sea_latitude, sea_longitude, start_date, end_date, sea_timezone)
```

Step 3: Load Data into pandas DataFrames

```python

import pandas as pd

df_lon = pd.DataFrame(lon_hourly)  # columns: ['time', 'precipitation']
df_sea = pd.DataFrame(sea_hourly)

df_lon.head(), df_sea.head()
```

Step 4: Data Cleaning (datetime index & missing values)

```python
# Step 4: Data Cleaning (datetime index & missing values)

# Convert to datetime and set index
df_lon['time'] = pd.to_datetime(df_lon['time'])
df_sea['time'] = pd.to_datetime(df_sea['time'])
df_lon.set_index('time', inplace=True)
df_sea.set_index('time', inplace=True)

# Ensure numeric and handle missing values
df_lon['precipitation'] = pd.to_numeric(df_lon['precipitation'], errors='coerce')
df_sea['precipitation'] = pd.to_numeric(df_sea['precipitation'], errors='coerce')

# Fill or drop missing (choose one strategy; here we fill with 0 since it's precipitation)
df_lon['precipitation'] = df_lon['precipitation'].fillna(0.0)
df_sea['precipitation'] = df_sea['precipitation'].fillna(0.0)

df_lon.info(), df_sea.info()
```

Step 5: Resample Hourly → Weekly Totals

```python
# Step 5: Resample Hourly → Weekly Totals

# Use weekly (W-SUN by default). If you want Monday-based weeks, use 'W-MON'.
weekly_lon = df_lon['precipitation'].resample('W-SUN').sum().rename('London (mm)')
weekly_sea = df_sea['precipitation'].resample('W-SUN').sum().rename('Seattle (mm)')

weekly_lon.head(), weekly_sea.head()
```

```python
# Step 6: Combine Weekly Series

weekly_combined = pd.concat([weekly_lon, weekly_sea], axis=1)
weekly_combined.head()
```

```python
# Step 7: Visualization — Weekly Total Precipitation

import matplotlib.pyplot as plt

ax = weekly_combined.plot(figsize=(12, 6), linewidth=2)
ax.set_title('Weekly Total Precipitation (mm) — London vs Seattle (2022)')
ax.set_xlabel('Week')
ax.set_ylabel('Total weekly precipitation (mm)')
ax.grid(True, alpha=0.3)
ax.legend(loc='upper right')
plt.show()
```

```python
# Step 8: Analysis — Totals & Simple Insights

total_lon = weekly_lon.sum()
total_sea = weekly_sea.sum()

winner = 'London' if total_lon > total_sea else 'Seattle'
diff = abs(total_lon - total_sea)

print(f"Total precipitation in 2022:")
print(f"  • London:  {total_lon:.1f} mm")
print(f"  • Seattle: {total_sea:.1f} mm")
print(f"\nOverall, {winner} had more precipitation by {diff:.1f} mm in 2022.")

# Optional: quick pattern checks
print("\nBasic pattern checks:")
print("  • Weeks with very high rain in London:\n", weekly_lon[weekly_lon > weekly_lon.quantile(0.9)])
print("\n  • Weeks with very high rain in Seattle:\n", weekly_sea[weekly_sea > weekly_sea.quantile(0.9)])
```

```python
# Step 9: Save Data to CSV (Optional)

weekly_combined.to_csv('weekly_precip_london_vs_seattle_2022.csv', index=True)
df_lon.to_csv('hourly_precip_london_2022.csv', index=True)
df_sea.to_csv('hourly_precip_seattle_2022.csv', index=True)
```
